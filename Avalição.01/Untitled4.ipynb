```python
import cv2                      # Biblioteca OpenCV (visão computacional)
import numpy as np              # Biblioteca para trabalhar com matrizes
from matplotlib import pyplot as plt  # Biblioteca para exibir imagens no Colab

img = cv2.imread('Debi.jpeg')  # Lê a imagem do arquivo (formato BGR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Converte BGR → RGB (necessário para exibir corretamente no matplotlib)

plt.imshow(img)                # Mostra a imagem
plt.axis('off')                # Remove os eixos (coordenadas)
plt.show()                     # Exibe a imagem na tela
```

# Teste 1 - Cinza
Função principal: simplificar a imagem (usado em quase todos os algoritmos)

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Converte imagem colorida para tons de cinza

plt.imshow(gray, cmap='gray')  # Mostra imagem em escala de cinza
plt.axis('off')
plt.show()
```

#Teste 2 - Blur(Suaviazação)
Função: reduzir ruídos e detalhes pequenos

```python
blur = cv2.GaussianBlur(img, (5,5), 0)
# Aplica um filtro de suavização usando uma janela 5x5

plt.imshow(blur)
plt.axis('off')
plt.show()
```

#Tete 3 - Binarização(threshold)
Função: transformar imagem em preto e branco (muito usado em detecção)

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Converte para cinza

_, binaria = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
# threshold: separa pixels em preto (0) e branco (255)
# 127 = valor limite (limiar)

plt.imshow(binaria, cmap='gray')
plt.axis('off')
plt.show()
```

#Teste 4 - Bordas (Canny)
Função: encontrar contornos e formas

```python
edges = cv2.Canny(gray, 100, 200)
# Detecta bordas na imagem
# 100 e 200 = limites inferior e superior de detecção

plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

#Teste 5 - Desenhar na Imagem
Função: marcar regiões na imagem

```python
img_copy = img.copy()
# Cria uma cópia da imagem original (para não alterar a original)

cv2.rectangle(img_copy, (50,50), (200,200), (255,0,0), 3)
# Desenha um retângulo
# (50,50) = ponto inicial
# (200,200) = ponto final
# (255,0,0) = cor (vermelho no RGB)
# 3 = espessura da linha

plt.imshow(img_copy)
plt.axis('off')
plt.show()
```

#Teste 6 - Alterar pixel
Função: modificar imagem pixel por pixel

```python
(b, g, r) = img[0, 0]  # lê pixel

img[0, 0] = (255, 0, 0)  # altera pixel para vermelho

plt.imshow(img)  # MOSTRA a alteração
plt.axis('off')
plt.show()
```

#Teste 7 - Alterar
Função: modificar imagem pixel por pixel

```python
for y in range(0, img.shape[0]):      # Percorre linhas
    for x in range(0, img.shape[1]):  # Percorre colunas
        img[y, x] = (255, 0, 0)       # Define cor do pixel
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img_resized = cv2.resize(img, (300, 300))

plt.imshow(img_resized)
plt.axis('off')
plt.show()
```
