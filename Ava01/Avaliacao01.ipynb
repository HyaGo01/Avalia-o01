4. (0,15 pt) Acesse as coordenadas centrais da imagem e exiba os valores R, G e B desse
pixel específico.

```python
altura, largura, _ = img_rgb.shape
centro_y, centro_x = altura // 2, largura // 2

pixel_central = img_rgb[centro_y, centro_x]
r, g, b = pixel_central

print(f"Coordenadas centrais: Y={centro_y}, X={centro_x}")
print(f"Valores no centro - R: {r}, G: {g}, B: {b}")
```

7. (0,15 pt) Separe os canais B, G e R da sua imagem utilizando funções do OpenCV

```python
b, g, r = cv2.split(img)

plt.figure(figsize=(10,3))

plt.subplot(1,3,1)
plt.imshow(b, cmap='gray')
plt.title('Canal Azul (B)')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.title('Canal Verde (G)')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(r, cmap='gray')
plt.title('Canal Vermelho (R)')
plt.axis('off')

plt.show()
```

13. (0,15 pt) Redimensione duas fotos para o mesmo tamanho e faça a mesclagem com
addWeighted (pesos 0.5/0.5).


```python
img2 = cv2.imread('Debi.png')
img2 = cv2.resize(img2, (img.shape[1], img.shape[0]))
blend = cv2.addWeighted(img, 0.5, img2, 0.5, 0)
plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB)); plt.axis('off')
```
